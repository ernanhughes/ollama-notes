```python
from sqlite3 import connect
from langchain_community.utilities import SQLDatabase
```

```python
db = SQLDatabase.from_uri("sqlite:///chinook.db", sample_rows_in_table_info = 3)
print(len(db.get_usable_table_names()), db.get_usable_table_names())
```

```python
print(db.table_info)
```

```python
cn = connect("chinook.db")
cur = cn.execute("Select Count(Distinct(AlbumId)) from Album;")
print(cur.fetchall())
```

```python
from langchain.llms import Ollama
from langchain_experimental.sql.base import SQLDatabaseChain
```

```python
llm = Ollama(model='llama3.2')
```

```python
db = SQLDatabase.from_uri("sqlite:///chinook.db")
db_chain = SQLDatabaseChain.from_llm(llm, db, verbose=True)
db_chain.run("How many Album are there?")
db_chain.run("How many Artist are there?")
```
